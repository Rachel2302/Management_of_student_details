Defined a function called insert() that inserts the user input data from the Tkinter GUI form into the Excel sheet.

3. The function then inserts the data from each input field into the appropriate cell in the next available row using the cell() method of the sheet object. The cell() method allows you to access a specific cell in the Excel sheet using row and column indices. The get() method of the input fields is used to retrieve the user input.
4. fter all the data has been inserted into the Excel sheet, the function saves the updated Excel sheet using the save() method of the wb (workbook) object.

<br>This part of the code is the main part of a Tkinter GUI application that allows users to enter student details and save them to an Excel sheet.<br>First it checks if the script is being run as the main module using the if __name__ == "__main__": statement.
1. If the script is the main module, a new Tkinter main window (root) is created using Tk().
2. The excel() function is called to initialize the Excel workbook and sheet. Labels for each input field are created using the Label widget and their properties are set (e.g., text, background color).
3. The labels are positioned in the main window using a grid layout with the grid() method. Entry fields for each input field are created using the Entry widget.
4. The Return event (pressing Enter key) is bound to the focus functions for each input field using the bind() method.
5. The entry fields are positioned in the main window using a grid layout with the grid() method, and the ipadx option is used to set the width of each input field.
6. A 'Submit' button is created using the Button widget, and its properties are set (e.g., text, foreground, and background colors). It is positioned in the main window using a grid layout with the grid() method.
7. The main event loop of the Tkinter application is started using the mainloop() method. This keeps the application running and waiting for user interactions.

```python
# Check if the script is being run as the main module

if __name__ == "__main__":
    # Create the main window (root) and set its properties
    root = Tk()
    root.configure(background='Yellow')
    root.title("Student Record")
    root.geometry("500x200")
    
    # Call the 'excel' function to initialize the Excel workbook and sheet
    excel()
    
    # Create labels for each input field and set their properties
    heading = Label(root, text="STUDENT DETAILS", bg="#ff7442")
    First_Name = Label(root, text="First Name", bg="#42ffd3")
    Last_Name = Label(root, text="Last Name", bg="#42ffd3")
    DOB = Label(root, text="DOB", bg="#42ffd3")
    Gender = Label(root, text="Gender", bg="#42ffd3")
    Contact_Number = Label(root, text="Contact Number", bg="#42ffd3")
    Email= Label(root, text="Email", bg="#42ffd3")
    Address = Label(root, text="Address", bg="#42ffd3")   
    Register_Number = Label(root, text="Register Number", bg="#42ffd3")
    Course = Label(root, text="Course", bg="#42ffd3")
    Major = Label(root, text="Major", bg="#42ffd3")
    GPA = Label(root, text="GPA", bg="#42ffd3")
    SSN_Number = Label(root, text="SSN Number", bg="#42ffd3")
    Technicial_Skills = Label(root, text="Technicial Skills", bg="#42ffd3")
    Soft_Skills = Label(root, text="Soft Skills.", bg="#42ffd3")
    Years_of_Experience = Label(root, text="Years of Experience", bg="#42ffd3")
    Number_of_Languages_Know = Label(root, text="Number of Languages Know", bg="#42ffd3")
    Hobbies = Label(root, text="Hobbies", bg="#42ffd3")
    
    # Position the labels in a grid layout
    heading.grid(row=0, column=1)
    First_Name.grid(row=1, column=0)
    Last_Name.grid(row=2, column=0)
    DOB.grid(row=3, column=0)
    Gender.grid(row=4, column=0)
    Contact_Number.grid(row=5, column=0)
    Email.grid(row=6, column=0)
    Address.grid(row=7, column=0)
    Register_Number.grid(row=8, column=0)
    Course.grid(row=9, column=0)
    Major.grid(row=10, column=0)
    GPA.grid(row=11, column=0)
    SSN_Number.grid(row=12, column=0)
    Technicial_Skills.grid(row=13, column=0)
    Soft_Skills.grid(row=14, column=0)
    Years_of_Experience.grid(row=15, column=0)
    Number_of_Languages_Know.grid(row=16, column=0)
    Hobbies.grid(row=17, column=0)
    
    # Create entry fields for each input field
    First_Name_field = Entry(root)
    Last_Name_field = Entry(root)
    DOB_field = Entry(root)
    Gender_field = Entry(root)
    Contact_Number_field = Entry(root)
    Email_field = Entry(root)
    Address_field = Entry(root)
    Register_Number_field = Entry(root)
    Course_field = Entry(root)
    Major_field = Entry(root)
    GPA_field = Entry(root)
    SSN_Number_field = Entry(root)
    Technicial_Skills_field = Entry(root)
    Soft_Skills_field = Entry(root)
    Years_of_Experience_field = Entry(root)
    Number_of_Languages_Know_field = Entry(root)
    Hobbies_field = Entry(root)
    
    # Bind the <Return> event (pressing Enter key) to the focus functions for each input field
    First_Name_field.bind("<Return>", focus1)
    Last_Name_field.bind("<Return>", focus2)
    DOB_field.bind("<Return>", focus3)
    Gender_field.bind("<Return>", focus4)
    Contact_Number_field.bind("<Return>", focus5)
    Email_field.bind("<Return>", focus6)
    Address_field.bind("<Return>", focus7)
    Register_Number_field.bind("<Return>", focus8)
    Course_field.bind("<Return>", focus9)
    Major_field.bind("<Return>", focus10)
    GPA_field.bind("<Return>", focus11)
    SSN_Number_field.bind("<Return>", focus12)
    Technicial_Skills_field.bind("<Return>", focus13)
    Soft_Skills_field.bind("<Return>", focus14)
    Years_of_Experience_field.bind("<Return>", focus15)
    Number_of_Languages_Know_field.bind("<Return>", focus16)
    Hobbies_field.bind("<Return>", focus17)
    
    # Position the entry fields in a grid layout
    First_Name_field.grid(row=1, column=1, ipadx="200")
    Last_Name_field.grid(row=2, column=1, ipadx="200")
    DOB_field.grid(row=3, column=1, ipadx="200")
    Gender_field.grid(row=4, column=1, ipadx="200")
    Contact_Number_field.grid(row=5, column=1, ipadx="200")
    Email_field.grid(row=6, column=1, ipadx="200")
    Address_field.grid(row=7, column=1, ipadx="200")
    Register_Number_field.grid(row=8, column=1, ipadx="200")
    Course_field.grid(row=9, column=1, ipadx="200")
    Major_field.grid(row=10, column=1, ipadx="200")
    GPA_field.grid(row=11, column=1, ipadx="200")  
    SSN_Number_field.grid(row=12, column=1, ipadx="200")
    Technicial_Skills_field.grid(row=13, column=1, ipadx="200")
    Soft_Skills_field.grid(row=14, column=1, ipadx="200")
    Years_of_Experience_field.grid(row=15, column=1, ipadx="200")  
    Number_of_Languages_Know_field.grid(row=16, column=1, ipadx="200")  
    Hobbies_field.grid(row=17, column=1, ipadx="200")  
    
    # Call the 'excel' function again to make sure the Excel sheet is properly initialized
    excel()
    
    # Create the 'Submit' button, set its properties, and bind it to the 'insert' function
    submit = Button(root, text="Submit", fg="Black", bg="#42ff46", command=insert)
    submit.grid(row=18, column=1)
    
    # Start the main event loop of the Tkinter application
    root.mainloop()
```
